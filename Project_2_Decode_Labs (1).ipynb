```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics (2).xlsx")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Product'].value_counts().plot(kind='bar')
```

```python
df.groupby('Product')['TotalPrice'].sum().plot(kind='bar')
```

```python
df['PaymentMethod'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
df['OrderStatus'].value_counts().plot(kind='bar')
```

```python
df['Quantity'].plot(kind='hist')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

Observations:
The dataset contains multiple products such as laptops, phones, tablets, etc., with varying sales distribution.
Some products have higher sales count, indicating higher demand among customers.
Revenue analysis shows that certain products contribute more to total sales compared to others.
The most commonly used payment methods are digital methods like Credit Card and Online payments.
Order status analysis indicates that most orders are successfully delivered, while a few are cancelled or returned.
The quantity distribution shows that most customers purchase fewer items per order.
Correlation analysis shows a positive relationship between quantity and total price, meaning higher quantity leads to higher revenue.
